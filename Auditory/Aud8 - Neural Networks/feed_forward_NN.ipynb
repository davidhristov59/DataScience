```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df = pd.read_csv('datasets/mushrooms.csv')
```

```python
df.head()
```

```python
encoder  = LabelEncoder()

for column in df.columns:
    df[column] = encoder.fit_transform(df[column].to_numpy().reshape(-1,1))
```

```python
x = df.drop(columns='class')
y = df['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
len(x_train.to_numpy()[0])
```

```python
x_test
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()

# vlezen sloj
model.add(Dense(32, input_dim = 22, activation = 'relu' ))

# hidden layer 
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))

model.add(Dense(1, activation='sigmoid')) #koristime sigmoid za binarna klasifikacija 
```

```python
df['class']
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=30 , batch_size= 32, validation_split=0.2)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred = y_pred.round()
```

```python
from sklearn.metrics import confusion_matrix, classification_report

confusion_matrix(y_test, y_pred)
```

```python
classification_report(y_test, y_pred)
```
