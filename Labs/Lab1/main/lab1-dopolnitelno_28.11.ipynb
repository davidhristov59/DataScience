```python
from selenium import webdriver
from selenium.webdriver.common.by import By # Automating a browser involves using software tools to control a web browser programmatically. This can include tasks such as navigating to web pages, filling out forms, clicking buttons, extracting data, and more.
import pandas as pd
import matplotlib.pyplot as plt
```

# Configuration

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

# Simple Usage

```python
# Initialize the webdriver.
browser = webdriver.Chrome(options=options)
```

```python
# Navigate to a web page.
browser.get('https://www.apple.com/mk/')
```

```python
# Print the page source code.
print(browser.page_source)
```

```python
# Select an element from the page.
element = browser.find_element(By.CLASS_NAME, 'headline')
```

```python
# Print the element's inner text.
print(element.text)
```

## Hints
- To navigate to a specific page with a given URL, use the `browser.get(url)` method.
- When selecting a single element, use `browser.find_element`, and for multiple elements, use `browser.find_elements`. Both of these functions take two arguments. The first specifies the attribute used to locate the element on the web page, and the second specifies the value of that attribute.
- The `By` class is used to specify which attribute is used to locate elements on a page. These are the various ways the attributes are used to locate elements on a page: `By.ID`, `By.NAME`, `By.TAG_NAME`, `BY.CLASS_NAME`, `By.CSS_SELECTOR`, etc.
- For more details, check the documentation (https://selenium-python.readthedocs.io/locating-elements.html).

## Solution

```python
driver = webdriver.Chrome(options=options)
```

```python
driver.get("https://books.toscrape.com/")
```

```python
print(driver.page_source)   
```

```python
book_titles = driver.find_elements(By.CSS_SELECTOR, '.product_pod h3 a')
book_titles
```

```python
book_prices = driver.find_elements(By.CSS_SELECTOR, '.product_price p:first-of-type')
book_prices
```

```python
book_titles = driver.find_elements(By.CSS_SELECTOR, '.product_pod h3 a')
book_prices = driver.find_elements(By.CSS_SELECTOR, '.product_price p:first-of-type')

products=[]

for i in range(len(book_titles)):
    
    dict = {
        "Book Title" : book_titles[i].get_attribute("title"),
        "Book Price" : book_prices[i].text.replace('£', '')
    }
    
    products.append(dict)

products
```

```python
df = pd.DataFrame(products)
df
```

```python
products_title_price_csv = df.to_csv('books-title-price.csv') 
```

```python
plt.hist(df['Book Price'], bins=30)
plt.xlabel('Price')
plt.ylabel('Count')
plt.xticks(rotation=315)
driver.quit()
```

```python
df = pd.read_csv('books-title-price.csv')
count_elements = len(df['Book Title'])
driver = webdriver.Chrome()
driver.get("https://books.toscrape.com/")

for i in range(20):
    book_url = driver.find_elements(By.CSS_SELECTOR, '.image_container a')
    url_of_a_book = book_url[i].get_attribute('href')
    driver.get(url_of_a_book)
    
     # Extract the stock information
    stock_text = driver.find_element(By.CSS_SELECTOR, ".instock").text
    stock_books = stock_text.split()[2].replace('(', '')

        # Assign the stock value to the DataFrame
    df.loc[i, 'Stock'] = int(stock_books)

        # Navigate back to the main page
    driver.get("https://books.toscrape.com/")

driver.quit()
```
